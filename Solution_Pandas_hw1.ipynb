```python
# Find ten most longest movies 
# movies_duration= df['duration']
# Function returns 10  largest mocies 
def Nmaxelements(movies_duration, N): 
    final_list = [] 
  
    for i in range(0, N):  
        max1 = 0
  
        for j in range(len(movies_duration)):      
            if movies_duration[j] > max1: 
                max1 = movies_duration[j]; 
                  
        movies_duration.remove(max1); 
        final_list.append(max1) 
          
    print(final_list) 
# Driver code 
movies_duration= df['duration'].tolist()
N=10
# Calling the function 
Nmaxelements(movies_duration, N) 
```
