```python
movie_crew_df = pd.read_csv('/Users/rakeshm/downloads/crew_data.tsv.gz', header = 0, delimiter = '\t', usecols=(0,1),
                            names = ['ID', 'director_ids'])
```

```python
person_df =pd.read_csv('/Users/rakeshm/downloads/name_data.tsv.gz', delimiter = '\t', header=0, 
                       usecols=(0,1), names=['person_ID', 'name'])
```

```python
person_df
```

```python
person_df.set_index('person_ID', inplace =True)
```

```python

#Merge movie_crew_df into df with IDs and drop the columns
pdd = pd.merge(df, movie_crew_df, on = 'ID', how = 'inner')
pdd.head()
#pdd.dropna(inplace=True)
```

```python
pdd.isna().sum()
```

```python
#Remove movies with more than one director and rename director_ids => director_id
pdd = pdd.loc[pdd["director_ids"].str.len()==9]
```

```python
pdd.head()
```

```python
pdd.rename(columns = {"director_ids":"director_id"}, inplace=True)
```

```python
pdd.head()
```

```python
#Merge df with person_df on director_id, remove director_id and person_ID columns, rename name => director

phd= pd.merge(pdd, person_df, left_on='director_id', right_on='person_ID', how='inner')
```

```python
phd.head()
```

```python
phd.drop(columns = 'director_id', axis=1, inplace=True)
phd.head()
```

```python
phd.rename(columns = {"name":"director"}, inplace=True)
```
